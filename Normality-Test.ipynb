# Normality Test on Customer Sales Data

This notebook demonstrates how to perform a normality test on a real-world dataset. We will focus on specific numerical columns to determine if their distributions follow a normal (Gaussian) distribution.

The process involves:
1.  **Loading the Data**: Using pandas to load the provided customer sales data.
2.  **Automated Testing**: Looping through all relevant numerical columns to perform visual and statistical tests.
3.  **Interpreting the Results**: Drawing conclusions based on the p-values from the tests for each column.

## 1. Setup and Data Loading

```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import io

# In a real scenario, you would load from a file like this:
df = pd.read_csv('Electronic-Sales.csv')

print("Data loaded successfully. Here are the first 5 rows:")
df.head()
```

## 2. Performing Normality Tests on All Numerical Columns

We will now loop through a list of specified numerical columns. For each column, we will generate a histogram and a Q-Q plot, then perform the Shapiro-Wilk and Kolmogorov-Smirnov tests.

The significance level (alpha) is set to 0.05. If the p-value is less than or equal to 0.05, we conclude the data is not normally distributed.

```python
columns_to_test = ['Age', 'Rating', 'Total Price', 'Unit Price', 'Quantity', 'Add-on Total']
sns.set_style('whitegrid')

for column in columns_to_test:
    print(f"\n--- Analysis for '{column}' ---\n")
    
    # --- Visual Inspection ---
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    fig.suptitle(f"Distribution Analysis for {column}", fontsize=16)
    
    # Histogram
    sns.histplot(df[column], kde=True, ax=axes[0], color='teal')
    axes[0].set_title('Histogram')
    
    # Q-Q Plot
    stats.probplot(df[column], dist="norm", plot=axes[1])
    axes[1].set_title('Q-Q Plot')
    
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
    
    # --- Statistical Tests ---
    data_column = df[column]
    
    # Shapiro-Wilk Test
    shapiro_stat, shapiro_p = stats.shapiro(data_column)
    print(f"Shapiro-Wilk Test:")
    print(f"Statistic: {shapiro_stat:.10f}, p-value: {shapiro_p:.10f}")
    if shapiro_p > 0.05:
        print("Conclusion: The data appears to be normally distributed.\n")
    else:
        print("Conclusion: The data does NOT appear to be normally distributed.\n")

    # Kolmogorov-Smirnov Test
    ks_stat, ks_p = stats.kstest(data_column, 'norm', args=(data_column.mean(), data_column.std()))
    print(f"Kolmogorov-Smirnov Test:")
    print(f"Statistic: {ks_stat:.10f}, p-value: {ks_p:.10f}")
    if ks_p > 0.05:
        print("Conclusion: The data appears to be normally distributed.")
    else:
        print("Conclusion: The data does NOT appear to be normally distributed.")
    print("--------------------------------------------------")
```

## 3. Final Conclusion

Based on both the visual inspections (skewed histograms and Q-Q plots deviating from the trend line) and the consistently low p-values (p <= 0.05) from the statistical tests, we can confidently conclude that **none of the analyzed numerical columns** (`Age`, `Rating`, `Total Price`, `Unit Price`, `Quantity`, `Add-on Total`) in this sample dataset are normally distributed.

This is a common finding in real-world data. For future analysis that requires a normality assumption (like a t-test), you would need to consider using non-parametric alternatives (like the Mann-Whitney U test) or attempt to transform the data.
